3) Describe your data cleaning approach. How and why did you decide on which cleaning techniques to use?

After establishing the data frame, the first step was to get the information about the data frame along with a description of the columns. The information was needed to see the number of null values, and the types of variables present to manipulate the data. A description showed if there were any obvious outliers, which there was in Price. That is why a histogram and boxplot was created to confirm it was present. Once identified, the data point was cleaned by removed. From there, I made sure there were no outliers in the float variables. 

The next course of action was about the null values. I created a data frame for each of the 4 columns of null values. To clarify, the data frame for price consisted of every price value being NaN, and the corresponding columns had their respective values for the row. The logic was to see if any NaN values were correlated to a specific column value. As an example, were an abnormally high number of price NaNs tied to a specific Region. However, this yielded no results. 

Now, I decided to check on the duplicates. Through coding it was discovered there were 50 duplicates, meaning the same Product, Region, Price, Units Sold, and Sales Date. This is not something that is going to happen, so I dropped them all.

At this point, apart from Sale Date each column had 100 values to fill. I chose to fill the Product missing values with “Other” and the Region one with “Missing”. The values must be filled in, and there’s not a statistical way to confidently fill those values with an actual name. Having an extra value is better than having incorrect ones. The Price and Units Sold columns were filled in differently. The Price values was completed by filling in the mean of the Price relative to the Product. This means for a Product that had multiple Prices and a NaN, the Price was filled by taking the mean of the Prices that the Product was connected to.  The Price of any item is related to its product, which is why the values were inputted this way. However, for Units Sold I created a histogram, and found that the distribution was even, which is why I filled in the missing values by taking the mean of the entire data set. I also felt that Units Sold is not related to a column enough to warrant inputting the values relative to another column like Price was.

Finally, now that there are no null values, I created a series of graphs to see if there were any outliers I wish to remove. While there is some Price values on these graphs that are fairly high, I did not believe they were isolated enough to be considered an outlier. As a result, the cleaning is done.
